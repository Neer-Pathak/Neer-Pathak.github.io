```python
from sklearn.tree import DecisionTreeRegressor
data_model = DecisionTreeRegressor(random_state=1)
```

# Using train test split

```python
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
train_x, val_x, train_y, val_y = train_test_split (x,y,random_state=0)
```

```python
data_model.fit (train_x,train_y)
```

```python
# getting predictions
prediction = data_model.predict(val_x)
print (mean_absolute_error(val_y, prediction ))
```

# Loop to compare different models using mean absolute error

```python
def get_mae(max_nodes, train_x,val_x, train_y, val_y):
    data_model = DecisionTreeRegressor(max_leaf_nodes=max_nodes,random_state=0)
    data_model.fit(train_x,train_y)
    predict = data_model.predict(val_x)
    mae = mean_absolute_error(val_y, predict)
    return (mae)
```

```python
for max_nodes in [5,50,500,5000]:
    mae_numbers = get_mae(max_nodes,train_x,val_x,train_y,val_y)
    print("max_leaf_nodes: %d \t\t Mean_absolute_error: %d" %(max_nodes,mae_numbers) )
```

# Using Random_forest_regressor

```python
from sklearn.ensemble import RandomForestRegressor
forest_model = RandomForestRegressor (random_state=1)
forest_model.fit(train_x,train_y)
for_preds = forest_model.predict(val_x)
print (mean_absolute_error(val_y,for_preds))
```
